### 01) WAP to find sum of all the elements in a List.

```python
lst = []
n = int(input("Enter number of elements: "))

for i in range(n):
    lst.append(int(input()))

total = 0
for i in lst:
    total = total + i

print("Sum =", total)
```

### 02) WAP to find largest element in a List.

```python
lst = []
n = int(input("Enter number of elements: "))

for i in range(n):
    lst.append(int(input()))

largest = lst[0]
for i in lst:
    if i > largest:
        largest = i

print("Largest =", largest)
```

### 04) WAP to interchange first and last elements in a list.

```python
lst = []
n = int(input("Enter number of elements: "))

for i in range(n):
    lst.append(input())

temp = lst[0]
lst[0] = lst[n-1]
lst[n-1] = temp

print("Updated List:", lst)
```

### 05) WAP to split the List into two parts and append the first part to the end.

```python
lst = []
n = int(input("Enter number of elements: "))

for i in range(n):
    lst.append(input())

print("list = ",lst)

k = int(input("Enter position: "))

new_list = []

for i in range(k, n):
    new_list.append(lst[i])

for i in range(k):
    new_list.append(lst[i])

print("Result:", new_list)
```

### 06) WAP to interchange the elements on two positions entered by a user.

### 07) WAP to reverse the list entered by user.

```python
lst = []

n = int(input("Enter number of elements: "))

for i in range(n):
    lst.append(int(input()))

rev = []
for i in range(n-1, -1, -1):
    rev.append(lst[i])

print("Reversed List:", rev)
```

### 11) WAP to print all odd numbers in a given range.

```python
start = int(input("Enter start: "))
end = int(input("Enter end: "))

oddL = []

for i in range(start, end + 1):
    if i % 2 != 0:
        oddL.append(i)

print("OddList = ",oddL)
```

### 12) WAP to count occurrences of an element in a list.

### 13) WAP to find second largest number in a list.

```python
lst = []
n = int(input("Enter number of elements: "))

for i in range(n):
    lst.append(int(input()))

largest = lst[0]
second = lst[0]

for i in lst:
    if i > largest:
        second = largest
        largest = i
    elif i > second and i != largest:
        second = i

print("Second Largest:", second)

    
```

### 14) WAP to extract elements with frequency greater than K.

```python
lst = []
n = int(input("Enter a Element: "))

for i in range(n):
    lst.append(int(input()))

ans = []

k = int(input("Enter a Frequency: "))

for i in range(n):
    count = 1;
    for j in range(i,n):
       
        if(lst[i]==lst[j]):
            count += 1
    if(count > k):
        ans.append(lst[i])
                
print("ans = ",ans)        
    
```

### 15) WAP to create a list of squared numbers from 0 to 9 with and without using List Comprehension.

### 16) WAP to create a new list (fruit whose name starts with 'b') from the list of fruits given by user.

```python
fruits = []
n = int(input("Enter number of fruits: "))

for i in range(n):
    fruits.append(input())

result = []
for f in fruits:
    if f[0] == 'b' or f[0] == 'B':
        result.append(f)

print("Filtered Fruits:", result)


```

### 17) WAP to create a list of common elements from given two lists.

```python
list1 = []
list2 = []

n1 = int(input("Enter size of first list: "))
for i in range(n1):
    list1.append(input())

n2 = int(input("Enter size of second list: "))
for i in range(n2):
    list2.append(input())

common = []

for i in list1:
    if (i in list2) and (i not in common):
        common.append(i)

print("Common Elements:", common)
```

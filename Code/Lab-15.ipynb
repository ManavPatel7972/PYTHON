### 03) WAP to create Circle class with area and perimeter function to find area and perimeter of circle.

```python
import math
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def circumference(self):
        return 2 * math.pi * self.radius
    
c1 = Circle(5)
print("Area of the circle:", c1.area())    
print("Circumference of the circle:", c1.circumference())
```

### 05) Create a bank account class with methods to deposit, withdraw, and check balance.

```python
class Bank_Acc:
    def __init__(self,acc_no, acc_name, acc_balance):
        self.acc_no = acc_no
        self.acc_name = acc_name
        self.acc_balance = acc_balance
        
    def deposit(self,amount):
        if(amount <=0):
            print("Balance should be >= 0")
            return
          
        else:
            self.acc_balance = self.acc_balance + amount

    def withdraw(self,amount):
        if(self.acc_balance >= amount):
            self.acc_balance -= amount
        else:
            print("Not enough Balance in your Account")
            return
         
    def check_balance(self):
        return self.acc_balance
            
b1 = Bank_Acc(101,"Manav", 1000)
print(b1.check_balance())

b1.deposit(1000)
print(b1.check_balance())

b1.withdraw(1000)
print(b1.check_balance())

b1.withdraw(500)
print(b1.check_balance())

        
```

### 06) Create a class for managing inventory that includes attributes such as item name, price, quantity, and methods to add, remove, and update items.

```python
class Inventory:

    def __init__(self):
        self.items = {} 

    def add_item(self, name, price, quantity):
        self.items[name] = {
            "price": price,
            "quantity": quantity
        }
        print(f"{name} added successfully.")


    def remove_item(self, name):
        if name in self.items:
            del self.items[name]
            print(f"{name} removed successfully.")
        else:
            print("Item not found.")


    def update_item(self, name, price=None, quantity=None):
        if name in self.items:
            if price is not None:
                self.items[name]["price"] = price
            if quantity is not None:
                self.items[name]["quantity"] = quantity
            print(f"{name} updated successfully.")
        else:
            print("Item not found.")

    def display_items(self):
        print("\n----- Inventory -----")
        # for name in self.items.items():
        #     print(name)
        #     print("Name" , name)
            # print("Details", details)
        for name, details in self.items.items():
            print(f"Item: {name}, Price: {details['price']}, Quantity: {details['quantity']}")


inv = Inventory()

inv.add_item("Pen", 10, 100)
inv.add_item("Notebook", 50, 40)

inv.display_items()

inv.update_item("Pen", price=12, quantity=120)

inv.remove_item("Notebook")

inv.display_items()
```

### 08) Create one class student_kit
#### Within the student_kit class create one class attribute principal name ( Mr ABC )
#### Create one attendance method and take input as number of days.
#### While creating student take input their name .
#### Create one certificate for each student by taking input of number of days present in class.

```python
class student_kit:


    principal_name = "Mr ABC"


    def __init__(self, name):
        self.name = name


    def attendance(self, days):
        self.days_present = days


    def generate_certificate(self):
        print("\n----- Certificate -----")
        print("Student Name:", self.name)
        print("Days Present:", self.days_present)
        print("Principal:", student_kit.principal_name)



    
n = int(input("Enter number of students: "))

students = []

for i in range(n):
    name = input("\nEnter student name: ")
    s = student_kit(name)

    days = int(input("Enter number of days present: "))
    s.attendance(days)

    students.append(s)

for s in students:
    s.generate_certificate()
```

### 09) Define Time class with hour and minute as data member. Also define addition method to add two time objects.

```python
class Time:


    def __init__(self, hour=0, minute=0):
        self.hour = hour
        self.minute = minute

    def add_time(self, t2):
        total_minutes = self.minute + t2.minute
        total_hours = self.hour + t2.hour

        extra_hour = total_minutes // 60
        total_minutes = total_minutes % 60
        total_hours += extra_hour

        return Time(total_hours, total_minutes)

    def display(self):
        print(f"{self.hour} hour(s) {self.minute} minute(s)")



h1 = int(input("Enter hour for Time 1: "))
m1 = int(input("Enter minutes for Time 1: "))
t1 = Time(h1, m1)

h2 = int(input("\nEnter hour for Time 2: "))
m2 = int(input("Enter minutes for Time 2: "))
t2 = Time(h2, m2)

result = t1.add_time(t2)

print("\n----- Result -----")
result.display()
```

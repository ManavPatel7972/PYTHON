### 01) WAP to find their union, intersection, and difference of the given two sets.
**Sample Input:**
```
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

```
**Sample Output:**
```
Union: {1, 2, 3, 4, 5, 6}
Intersection: {3, 4}
Difference (A-B): {1, 2}

```

````python
# A = {1, 2, 3, 4}
# B = {3, 4, 5, 6}

# print("Union:", A | B)
# print("Intersection:", A & B)
# print("Difference (A-B):", A - B)

A = set(map(int, input("Enter elements of A ").split()))
B = set(map(int, input("Enter elements of B ").split()))

print("Union:", A | B)
print("Intersection:", A & B)
print("Difference (A-B):", A - B)

````

### 05) WAP to Merge two dictionaries. If a key exists in both dictionaries, add their values.

**Sample Input:**
```
Dict1 = {1: 10, 2: 20, 3: 30}
Dict2 = {2: 40, 3: 50, 4: 60}

```
**Sample Output:**
```
{1: 10, 2: 60, 3: 80, 4: 60}
```

````python
# lst1 = tuple(int,map(list(map(int, input("Enter list elements: ").split()))).split(","))

# print(lst1)

# for i in Dict2:
#     Dict1[i]  = Dict1.get(i,0) + Dict2[i]

# print(Dict1)


d1 = {}
n1 = int(input("Enter elements of d1: "))
for i in range(n1):
    k, v = input().split()
    d1[k] = int(v)

d2 = {}
n2 = int(input("Enter elements of d2: "))
for i in range(n2):
    k, v = input().split()
    d2[k] = int(v)

for k in d2:
    d1[k] = d1.get(k, 0) + d2[k]

print( d1)


````

### 07) WAP to find the keys having maximum and minimum values in a dictionary.

**Sample Input:**
```
{'A': 50, 'B': 90, 'C': 30}
```
**Sample Output:**
```
Max Value Key: B
Min Value Key: C

```

````python
d = {}
n = int(input("Enter number of elements: "))

for i in range(n):
    k, v = input().split()
    d[k] = int(v)

print("Max Key:", max(d, key=d.get))
print("Min Key:", min(d, key=d.get))
````

### 09) WAP to count the frequency of each character in a string using a dictionary.
**Sample Input:**
```
programming
```
**Sample Output:**
```
{'p': 1, 'r': 2, 'o': 1, 'g': 2, 'a': 1, 'm': 2, 'i': 1, 'n': 1}
```

````python
s = input("Enter string: ")
f = {}

for ch in s:
    f[ch] = f.get(ch, 0) + 1

print(f)
````

### 10) WAP to swap keys and values of a dictionary.
**Sample Input:**
```
{1: 'A', 2: 'B', 3: 'C'}
```
**Sample Output:**
```
{'A': 1, 'B': 2, 'C': 3}
```

````python
d = {}
n = int(input("Enter number of elements: "))

for i in range(n):
    k, v = input().split()
    d[k] = v

print({v: k for k, v in d.items()})

````

### 11) WAP to find the student with the highest total marks, given student names and their marks.

**Sample Input:**
```
{'Amit': [70, 80, 90], 'Riya': [85, 75, 80], 'Neha': [60, 70, 65]}
```
**Sample Output:**
```
Topper: Amit
```

````python
students = {}
n = int(input("Enter number of Ele: "))

for i in range(n):
    name = input("Name: ")
    marks = list(map(int, input("Enter marks: ").split()))
    students[name] = marks

def getMax(item):
    return item[1]

top = max(students, key=students.get)
print("Top Student:", top)
````

### 12) WAP to extract keys whose values are even numbers.

**Sample Input:**
```
{1: 11, 2: 20, 3: 15, 4: 40}
```
**Sample Output:**
```
[2, 4]
```

````python
d = {}
n = int(input("Enter number of elements: "))

for i in range(n):
    k, v = input().split()
    d[k] = int(v)

print("Even Value Keys:", [int(k) for k in d if d[k] % 2 == 0])
````

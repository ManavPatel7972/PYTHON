### 01) Write a function to calculate BMI given mass and height. (BMI = mass/h**2)

```python
mass = float(input("Enter a mass:"))
height = float(input("Enter a height:"))
def bmi(mass, height):
    return mass / (height ** 2)

print(f"BMI = {bmi(mass, height)}")
```

### 02) Write a function that add first n numbers.

```python
n = int(input("Enter a n:"))
def sumOfN(n):
    total = 0
    for i in range(1, n+1):
        total += i
    return total

print("Sum of n is = ",sumOfN(n))
```

### 03) Write a function that returns 1 if the given number is Prime or 0 otherwise.

```python
n = int(input("Enter a n:"))

def isPrime(n):
    if n <= 1:
        return 0
    for i in range(2, n):
        if n % i == 0:
            return 0
    return 1

print(isPrime(n))

```

### 04) Write a function that returns the list of Prime numbers between given two numbers.

```python
n1 = int(input("Enter a n1:"))
n2 = int(input("Enter a n2:"))

def primeRange(a,b):
    primes = []
    for num in range(a, b+1):
        if num > 1:
            for i in range(2, num):
                if num % i == 0:
                    break
            else:
                primes.append(num)
    return primes

print('Prime List = ',primeRange(n1,n2))
```

### 05) Write a function that returns True if the given string is Palindrome or False otherwise.

```python
s = input("Enter a String:")
def is_palindrome(s):
    return s == s[::-1]

print(is_palindrome(s))
```

### 08) Write a recursive function to find nth term of Fibonacci Series.

```python
n = int(input("Enter a number n: "))

def fibbo(n):
    if n < 0:
        return -1
    if n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        return fibbo(n-1) + fibbo(n-2)

print("Nth term is = ", fibbo(n))


```

### 10) Write a function to get the sum of the scores ending with zero.
#### Example : scores = [200, 456, 300, 100, 234, 678] 
#### Ans = 200 + 300 + 100 = 600

```python
def sumEndZero(scores):
    total = 0
    for s in scores:
        if s % 10 == 0:
            total += s
    return total

n = int(input("Enter a number of Element:"))
lst = []

for i in range(n):
    lst.append(int(input()))


r = sumEndZero(lst)
print("sum of the scores ending with zero = ", r)
```

### 11) Write a function to invert a given Dictionary.
#### hint: keys to values & values to keys
#### Before : {'a': 10, 'b':20, 'c':30, 'd':40}
#### After : {10:'a', 20:'b', 30:'c', 40:'d'}

```python
def invert_dict(d):
    dict1 = {}
    for key, value in d.items():
        dict1[value] = key
    return dict1


dict1 = {'a': 10, 'b': 20, 'c': 30, 'd': 40}

r = invert_dict(dict1)
print("Inverted Dictionary:", r)
```

### 12) Write a function that returns the number of uppercase and lowercase letters in the given string. 
#### example : Input : s1 = AbcDEfgh ,Ouptput : no_upper = 3, no_lower = 5

```python
def countChar(s):
    upper = 0
    lower = 0

    for ch in s:
        if ch.isupper():
            upper += 1
        elif ch.islower():
            lower += 1

    return upper, lower

s1 = "AbcDEfgh"

u, l = countChar(s1)
print("Uppercase letters = ", u)
print("Lowercase letters = ", l)
```

### 13) Write a lambda function to get smallest number from the given two numbers.

```python
n1 = int(input("Enter a n1:"))
n2 = int(input("Enter a n2:"))

small = lambda n1, n2: n1 if n1 < n2 else n2

print("Smallest number:", small(n1, n2))
```

### 14) For the given list of names of students, extract the names having more that 7 characters. Use filter().

```python
# lst = ["adhsjjdjed", "axscshcsb","scsd","hwdhefherfkef","dcdd","xsdff"]

students = ["Rahul", "Amitabh", "Siddharth", "Neha", "Kunal", "Priyanka", "Abhishek"]

names = list(filter(lambda name: len(name) > 7, students))

print(names)
```

### 15) For the given list of names of students, convert the first letter of all the names into uppercase. use map().

```python
students = ["rahul", "amitabh", "siddharth", "neha", "kunal", "priyanka","Abhishek"]

capitalizeNames = list(map(lambda name: name.capitalize(), students))

print(capitalizeNames)
```

### 16) Write udfs to call the functions with following types of arguments:
1. Positional Arguments
2. Keyword Arguments
3. Default Arguments
4. Variable Legngth Positional(*args) & variable length Keyword Arguments (**kwargs)
5. Keyword-Only & Positional Only Arguments

```python
def add(a, b):
    print("Sum =", a + b)

add(10, 20)

def student(name, age):
    print("Name:", name)
    print("Age:", age)

student(age=20, name="Rahul")


def greet(name="Student"):
    print("Hello", name)

greet()
greet("Amit")


def total(*numbers):
    print("Sum =", sum(numbers))

total(10, 20)
total(5, 10, 15, 20)


def student_info(**details):
    for key, value in details.items():
        print(key, ":", value)

student_info(name="Ravi", age=21, city="Rajkot")


def add(a, b, /):
    print(a + b)

add(10, 20)

def info(*, name, age):
    print(name, age)

info(name="Kiran", age=22)
```

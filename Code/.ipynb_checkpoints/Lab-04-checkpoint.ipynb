### 04) WAP to print numbers between two given numbers which is divisible by 2 but not divisible by 3.

```python
start = int(input("Enter starting number: "))
end = int(input("Enter ending number: "))

print("Numbers divisible by 2 but not divisible by 3 are:")

for i in range(start, end + 1):
    if i % 2 == 0 and i % 3 != 0:
        print(i)
```

### 06) WAP to print sum of series 1 + 4 + 9 + 16 + 25 + 36 + ...n.

```python
n = int(input("Enter the value of n: "))
sum = 0
for i in range(1,n+1):
    sum += i**2
    print(f"{i**2} +", end=" ")
print(f"\nSum of series = {sum}")
```

### 07) WAP to print sum of series 1 – 2 + 3 – 4 + 5 – 6 + 7 ... n.

```python
n = int(input("Enter the value of n: "))
sum = 0

for i in range(1,n+1):
    if(i%2!=0):
        print(f" {i} ",end="-")
        sum += i
    else:
        print(f" {i} ",end="+")
        sum -= i
print(f"\nSum of Series = {sum}")

```

### 10) WAP to print GCD of given two numbers.

```python
n1 = int(input("Enter the value of n1: "))
n2 = int(input("Enter the value of n2: "))

min = n1 if n1<n2 else n2
gcd = 1

for i in range(1,min+1):
    if(n1 % i == 0  and n2 % i == 0):
        gcd = i

print(f"GCD = {gcd}")
        
# METHOD 2:
# import math

# gcd = math.gcd(n1,n2)
# print(f"gcd = {gcd}")
```

### 12) WAP to find whether the given number is Prime or not.

```python
import math
n = int(input("Enter the value of n: "))
flag = False
if(n == 1 ):
    print(f"{n} is Prime")
    
for i in range(2,int(math.sqrt(n+1))):
    if(n % i == 0):
        flag = True
        break;

if(flag == True):
    print(f"{n} is Not Prime.")
else:
    print(f"{n} is Prime.")
        
```

### 14) WAP to check whether the given number is Palindrome or not.

```python
n = int(input("Enter the value of n: "))
rev = list(str(n))
check = []

print(rev)
for i in range(len(rev)-1,-1,-1):
    print(rev[i])
    check.append(str(rev[i]))

print(rev)
print(check)
if(check == rev ):
    print(f"{n} is palindrome.")
else:
    print(f"{n} is not palindrome.")


# METHDO 2:
n = str(n)
reverse = n[::-1]
print(reverse)
    
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
num = int(input("Enter a number: "))
sum = 0
digi = len(str(num))

temp = num
while temp > 0:
   digit = temp % 10
   sum += digit ** digi
   temp //= 10

if num == sum:
   print(num,"is an Armstrong number")
else:
    print(num,"is Not Armstrong number")
```

### 05) WAP to generate 3 random integers between 100 and 999 which is divisible by 5.

```python
import random

count = 0
li = []

while count < 3:
    num = random.randint(100, 999)
    
    if num % 5 == 0:
        print(li.append(num))
        count += 1
print("list = ", li)
```

### 06) WAP to generate 100 random lottery tickets and pick two lucky tickets from it and announce them as Winner and Runner up respectively.

```python
import random

li = []

for i in range(100):
    ticket = random.randint(100000, 999999)
    li.append(ticket)


res = random.sample(li, k=2)
print("res = ", res)

print("W1", res[0])
print("W2", res[1])
```

### 08) Subtract a week (7 days) from a given date in Python.

```python
from datetime import datetime, timedelta

li = list(input("Enter a Date (YYYY-MM-DD) :").split("-"))
print(li)

year, month, day = li

given_date = datetime(int(year), int(month), int(day))

new_date = given_date - timedelta(days=7)

print("new_date = ", new_date)
```

### 09) WAP to Calculate number of days between two given dates.

```python
from datetime import datetime

date1 = input("Enter first date (DD-MM-YYYY): ")
date2 = input("Enter second date (DD-MM-YYYY): ")

d1 = datetime.strptime(date1, "%d-%m-%Y")
d2 = datetime.strptime(date2, "%d-%m-%Y")

difference = abs((d2 - d1).days)

print("Number of days between two dates:", difference)
```

### 10) WAP to Find the day of the week of a given date.(i.e. wether it is sunday/monday/tuesday/etc.)

```python
from datetime import datetime

date_input = input("Enter date (DD-MM-YYYY): ")

given_date = datetime.strptime(date_input, "%d-%m-%Y")

day_name = given_date.strftime("%A")

print("Day of week = ", day_name)
```

### 03) WAP to find tuples which have all elements divisible by K from a list of tuples.
**Sample Input:**
```
tuple_list = [(2, 4, 6), (3, 6, 9), (4, 8, 12), (5, 10, 15)]
K = 2

```
**Sample Output:**
```
Tuples divisible by 2 : [(2, 4, 6), (4, 8, 12)]
```

````python
n = int(input("Enter number of tuples: "))
lst = []

for i in range(n):
    lst.append(tuple(map(int, input("Enter tuple elements: ").split())))

k = int(input("Enter K: "))

result = []
for t in lst:
    if all(x % k == 0 for x in t):
        result.append(t)

print("Tuples divisible by", k, ":", result)
````

### 14) WAP to create (index, value) pairs for prime numbers.

**Sample Input:**
```
T = (4, 7, 9, 11, 15)
```
**Sample Output:**
```
((1, 7), (3, 11))
```

````python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

t = tuple(map(int, input("Enter tuple elements: ").split()))

result = []
for i in range(len(t)):
    if is_prime(t[i]):
        result.append((i, t[i]))

print("Prime index-value pairs:", tuple(result))
````

### 16) WAP to compute column-wise sum from a tuple of lists.

**Sample Input:**
```
T = ([1, 2, 3], [4, 5, 6], [7, 8, 9])
```

**Sample Output:**
```
12 15 18

```

````python
t = int(input("Enter number of tuple: "))
data = []

for i in range(t):
    data.append(list(map(int, input("Enter row elements: ").split())))

for col in zip(*data):
    print(sum(col), end=" ")
````
